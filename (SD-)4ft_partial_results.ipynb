Lets order categories in each delay atributes so then we can use lcut or rcut.

```python
# Desired sorting order
desired_order = ['Low', 'Noticeable', 'Medium', 'Above Average', 'High', 'Very High', 'Extreme']

# Columns to apply the ordering
columns_to_order = ['CARRIER_DELAY_category', 'WEATHER_DELAY_category', 'NAS_DELAY_category', 'SECURITY_DELAY_category', 'LATE_AIRCRAFT_DELAY_category']

# Convert the columns to a categorical data type with the desired order
for column in columns_to_order:
    df[column] = pd.Categorical(df[column], categories=desired_order, ordered=True)

# Verify the result for one of the columns
print(df['WEATHER_DELAY_category'].unique())
```

```python
# Verify the result for one of the columns
print(df['SECURITY_DELAY_category'].unique())
```

# Lets experiment with different procedures and see first results.

```python
clm = cm(df=df,proc='4ftMiner',
               quantifiers= {'aad':0.1, 'conf' : 0.02, 'RelBase':0.001},
               ante ={
                    'attributes':[
                        {'name': 'OP_UNIQUE_CARRIER', 'type': 'subset', 'minlen': 1, 'maxlen': 1}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
               succ ={
                    'attributes':[
                        {'name': 'CARRIER_DELAY_category', 'type': 'rcut', 'minlen': 1, 'maxlen': 4},
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
              cond ={
                    'attributes':[
                        {'name': 'DEST', 'type': 'one', 'value' : 'JFK'},
                        {'name': 'DEST', 'type': 'one', 'value' : 'LGA'}
                    ], 'minlen':1, 'maxlen':1, 'type':'con'},
              opts = {'max_categories' : 500}
               )


print(clm.result)
clm.print_summary()
clm.print_rulelist()
clm.print_rule(1)
clm.print_rule(3)
```
